- crosstab


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df
```

Pegar a quantidade de pedidos por cidade (city) e categoria (category):

```python
pd.crosstab(df.City, df.Category)
```

Também podemos criar um crosstab com um subtotal por linhas:
- margins: True;

```python
pd.crosstab(df.City, df.Category, margins=True)
```

Podemos também adicionar níveis:
- No exemplo abaixo ele mostra a categoria do produto e a forma de envio das mesmas por cidade:

```python
pd.crosstab(df.City, [df.Category, df['Ship Mode']], margins=True)
```

Também podemos concatenar valores:
- Abaixo, todas as sales de cada categoria por cidade foram somadas:



```python
pd.crosstab([df.City], [df.Category], values=df.Sales, aggfunc=(np.sum), margins=True)
```

Função Normalize

Uso do groupby:
- Agrupar as colunas do nosso dataframe;
- Abaixo, agrupando por categoria:

```python
df_type_furniture = df.groupby('Category')
df_type_furniture
```

Consigo ver a quantidade de grupos que temos:

```python
df_type_furniture.ngroups
```

Não é necessário fazer todos esses passos separadamente, é possível fazer tudo junto:

```python
df_type_furniture = df.groupby('Category').ngroups
```

Também posso mostrar os grupos que você tem e as linhas que compõem esse atributo:

```python
df.groupby('Category').groups
```

Atributo size:
- Para saber a quantidade de registros:

```python
df.groupby('Category').size()
```

Atributo get_group:
- Usado para filtrar um grupo específico e o tamanho do seu dataframe.

```python
df.groupby('City').get_group('Akron')
```

```python
df.groupby('City').mean()
```

Também podemos agrupar por mais de um campo:

```python
df.groupby(['City', 'Category'])['Sales'].sum()
```

Append

> Uma outra forma de concatenar.

- Se os dataframes foram de mesmo tamanho e as colunas forem iguais, ele vai encaixar um embaixo do outro.
- Se não forem, ou as colunas forem diferentes, o método faz a colocação do NaN.

df_append = df1.append(df2)

Também é possível:

df_append = df1.append(df2, ignore_index=True)
