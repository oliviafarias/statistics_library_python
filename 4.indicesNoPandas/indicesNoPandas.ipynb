Loc
> Com o loc conseguimos pegar os registros que necessita dentro do dataframe. O pandas nos fornece esse comando para pegar uma determinada coluna pelo nome.

> loc[:,] - O loc espera dois parâmetros. O primeiro são as linhas e o segundo a coluna.





```python
import pandas as pd
df_prouni = pd.read_excel('cursos-prouni.csv.xlsx')
df_prouni
```

```python
df_medicina = df_prouni[df_prouni['uf_busca'] == 'SP'	]
df_medicina
```

Traga todas as linhas da coluna mensalidade:

```python
df1 = df_medicina.loc[:, 'mensalidade']
df1
```

Traga desde a linha 41438 até a linha 41443, e traga a coluna 'mensalidade':

```python
df2 = df_medicina.loc[41438:41443, ['curso_busca', 'mensalidade', 'turno']]
df2
```

iloc:
> A diferença entre o loc e o iloc: O loc são as locations, você consegue trazer as informações pelo nome da coluna (quando ela está nomeada), por exemplo, mas quando você tem um dataset que não está nomeado, e as colunas possuem um nome default, usa-se o iloc (pelo índice da coluna).

> Ambos são bem parecidos:
  * O primeiro parâmetro são as linhas que se quer retornar
  * E no segundo parâmetro as colunas.

A diferença é que ao invés de colocar o nome da coluna, coloca-se o índice.




Me traga todas as linhas, coluna 1:

```python
df1 = df_medicina.iloc[:, 1]
df1
```

Retorne a primeira linha do dataframe:

```python
df2 = df_medicina.iloc[0]
df2
```

Retorne a informação do df_medicina da linha 0 a linha 1, de todas as colunas:

```python
df3 = df_medicina.iloc[0:4, :]
df3
```

Retorne todas as linhas, da coluna 2 até a 4:

```python
df4 = df_medicina.iloc[:, 2:4]
df4
```

Retorne a linha 1 a 5, das colunas 2 a 7:

```python
df5 = df_medicina.iloc[1:5, 2:7]
df5
```

Reset Índices:

> reset_index



```python
df_medicina.reset_index()
```

Quando criado dessa forma, ele cria uma outra coluna com o índice antigo. Para que isso não ocorra, deve-se passar como parâmetro para tirar essa coluna:
- reset_index(drop=True)

E para que ele mantenha esse novo índice:
- reset_index(inplace=True, drop=True)

```python
df_medicina.reset_index(inplace=True, drop=True)
df_medicina
```
