Map e Replace:

> Map: A função Map substitui valores por outros; Na hora de passar o comando, o map recebe como parâmetro de entrada três tipos:
  - Uma series;
  - Uma função ou
  - Um dicionário;

> Replace: A diferença entre o map e o replace é que o map aceita series, funções, dicionários, enquanto o replace aceita além disso tudo:
  - Números (inteiros, float);
  - Strings;
  - Regex;
  - E mais coisas como dados de entrada;







Criando uma nova coluna no DataFrame:

- Passa-se o nome do dataframe e o nome da coluna;
- Chamo o dataframe original (data)
- Passa-se as colunas que se quer utilizar
- map(o que quero substituir)
  "passei um dicionário - procura todo mundo que é zero e transforma em 2.00""


```python
df['new_dividends'] = data['Dividends'].map({0:2.00})
```

```python
df
```

Foram feitar alterações para todas as linhas da coluna, mas é possível usar algum tipo de filtro/condição pra que só seja feita a alteração se a condição for TRUE.
  - where (onde o index de data seja...)

```python
df['new_dividends'] = data['Dividends'].map({0:2.00}).where(data.index == '2022-10-31')
df
```

Dados Faltantes
  Função dentro do pandas para descobrir se temos dados faltantes:
  - isNull;

```python
df_missing = pd.DataFrame({'A' : [0, 1, 2, None, 4],
                           'B' : [5, 6, None, 8, 9],
                           'C' : ['a', 'b', 'a', 'a', None]})
df_missing
```

```python
df_missing.isnull()
```

```python
df_missing.isnull().sum()
```

```python
media = df_missing['A'].mean()
df_missing = df_missing['A'].fillna(media)
df_missing
```

fillna = Comando para preencher o que é nulo;
