A seguir, vamos agrupar os campos uf_busca e curso_busca e agregar uma média da mensalidade.
A função agg tem de se passar o que se quer realizar (no caso, uma média dos valores de mensalidade).

```python
df_prouni.groupby(['uf_busca', 'curso_busca'])['mensalidade'].agg(['mean'])
```

```python
df_prouni.groupby(['uf_busca', 'curso_busca'])['mensalidade'].agg(['mean']).sort_values(by=['uf_busca', 'mean'])
```

```python
def duplicaBolsa(rows):
  nova_bolsa = rows * 2
  return nova_bolsa
```

```python
df = df_prouni[['uf_busca', 'curso_busca', 'mensalidade', 'bolsa_integral_ampla']]
df
```

```python
df['nova_bolsa'] = df['bolsa_integral_ampla'].apply(duplicaBolsa)
```

```python
df
```

```python
df['nova_bolsa'] = df['nova_bolsa'].fillna(1)
```

```python
df
```

A linguagem Python possui a palavra reservada 'lamba' que é uma função anônima.
* Para funções simples e rápidas;

o que seria:
def duplica_bolsa(rows):
  nova_bolsa = rows * 2

=

lambda x : x * 2
  -> rows : rows * 2

  

```python
df['nova_bolsa_lambda'] = df['bolsa_integral_ampla'].apply(lambda x: x*2)
```

```python
df
```

Uma outra forma:

```python
df = df.assign(nova_bolsa_lambda2 = lambda x : (x['bolsa_integral_ampla'] * 2))
df
```
